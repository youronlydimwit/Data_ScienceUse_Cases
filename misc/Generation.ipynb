```python
import numpy as np
import pandas as pd
```

```python
# Generating random data with a normal distribution
data = np.random.normal(loc=3, scale=1, size=(500, 12))  # mean=3, standard deviation=1

# Rounding the values and ensuring they are between 1 and 5
data = np.round(data)
data[data < 1] = 1
data[data > 5] = 5

# Converting to integers
data = data.astype(int)

# Creating a DataFrame
df = pd.DataFrame(data, columns=[f'Column_{i}' for i in range(1, 13)])

# Displaying the DataFrame
df.head(5)
```

```python
import matplotlib.pyplot as plt

# Setting up the subplots
fig, axes = plt.subplots(3, 4, figsize=(15, 10))
fig.suptitle('Histograms for Each Column')

# Visualizing/histogram for each column
for i, ax in enumerate(axes.flat):
    column = df.columns[i]
    ax.hist(df[column], bins=[1, 2, 3, 4, 5, 6], alpha=0.5, edgecolor='black')
    ax.set_title(f'{column}')
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Selecting random columns
skew_left = np.random.choice(df.columns, 3, replace=False)

# Introducing skewness to the selected columns
for column in skew_left:
    skewness_factor = np.random.uniform(0.1, 0.5)  # Random skewness factor between 0.1 and 0.5
    df[column] -= int(skewness_factor * 4)  # Shifting values towards 1

# Displaying the modified DataFrame
df.head(5)
```

```python
# Setting up the subplots
fig, axes = plt.subplots(3, 4, figsize=(15, 10))
fig.suptitle('Histograms for Each Column')

# Visualizing/histogram for each column
for i, ax in enumerate(axes.flat):
    column = df.columns[i]
    ax.hist(df[column], bins=[1, 2, 3, 4, 5, 6], alpha=0.5, edgecolor='black')
    ax.set_title(f'{column}')
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Selecting random columns for right skewness, excluding the ones already skewed left
skew_right = np.random.choice([col for col in df.columns if col not in skew_left], 2, replace=False)

# Introducing skewness to the selected columns
for column in skew_right:
    skewness_factor = np.random.uniform(0.1, 0.5)  # Random skewness factor between 0.1 and 0.5
    df[column] += int(skewness_factor * 4)  # Shifting values towards 5

# Displaying the modified DataFrame
df.head(5)
```

```python
# Setting up the subplots
fig, axes = plt.subplots(3, 4, figsize=(15, 10))
fig.suptitle('Histograms for Each Column')

# Visualizing/histogram for each column
for i, ax in enumerate(axes.flat):
    column = df.columns[i]
    ax.hist(df[column], bins=[1, 2, 3, 4, 5, 6], alpha=0.5, edgecolor='black')
    ax.set_title(f'{column}')
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import random

# Function to generate staff ID
def generate_staff_id():
    level_codes = ['DR'] * 3 + ['MA'] * 50 + ['SP'] * 75 + ['SA'] * 372  # Level codes distribution
    level_code = random.choice(level_codes)  # Randomly choose a level code
    random_numbers = ''.join(str(random.randint(0, 9)) for _ in range(5))  # Generate 5 random numbers
    return f"{level_code}{random_numbers}"

# Add "Staff_Id" column to DataFrame
df['Staff_Id'] = [generate_staff_id() for _ in range(500)]

# Display the DataFrame with the new "Staff_Id" columns
df.head(5)
```

```python
# Generating random values for Month_Of_Service
df['Month_Of_Service'] = [random.randint(0, 66) for _ in range(500)]  # 66 months = 5 years 6 months

# Generating Years_Of_Service based on Month_Of_Service
df['Years_Of_Service'] = df['Month_Of_Service'] // 12  # Integer division to get years

# Adjusting Years_Of_Service for people with less than a year of service
df.loc[df['Years_Of_Service'] == 5, 'Years_Of_Service'] = 4

# Displaying the DataFrame with the new columns
df.head(5)
```

```python
# Define the possible residence locations
residence_locations = ['Jakarta', 'Tangerang', 'Bekasi', 'Depok', 'Bogor']

# Generating random values for Residence
df['Residence'] = [random.choice(residence_locations) for _ in range(500)]

# Creating Residence_Code based on Residence
residence_mapping = {location: i+1 for i, location in enumerate(residence_locations)}
df['Residence_Code'] = df['Residence'].map(residence_mapping)

# Displaying the DataFrame with the new columns
df.head(5)
```

```python
# Define salary ranges for each staff level
salary_ranges = {'SA': (5070000, 7004030),  # Salary range for Staff (SA)
                 'SP': (8100075, 10240060), # Salary range for Supervisor (SP)
                 'MA': (15562000, 21053011), # Salary range for Manager (MA)
                 'DR': (53010000, 55020000)} # Salary range for Director (DR)

# Function to generate net salary based on staff level
def generate_net_salary(level_code):
    lower_bound, upper_bound = salary_ranges[level_code]
    return random.randint(lower_bound, upper_bound)

# Add "Net_Salary" column to DataFrame
df['Net_Salary'] = [generate_net_salary(staff_id[:2]) for staff_id in df['Staff_Id']]

# Display the DataFrame with the new "Net_Salary" column
df.head(5)
```

```python
# Grouping by staff level and calculating median net salary
median_salary_by_level = df.groupby(df['Staff_Id'].str[:2])['Net_Salary'].median()

# Displaying the median net salary for each staff level
print(median_salary_by_level)
```
